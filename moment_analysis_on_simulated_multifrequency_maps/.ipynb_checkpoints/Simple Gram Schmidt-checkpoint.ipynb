```python
def gram_schmidt(vectors):
    basis = []
    for v in vectors:
        w = v - np.sum( np.dot(v,b)*b  for b in basis )
        if (w > 1e-10).any():  
            basis.append(w/np.linalg.norm(w))
    return np.array(basis)
```

```python
x=linspace(10,3000)
y1=x ; y2=x**2. ; y3=x**3. ; y4=x**4. ; y5=x**5.
vectors=[y1,y2,y3,y4,y5]
```

```python
basis=gram_schmidt(vectors)
```

```python
figure()
for i in range(5):
    plot(x,basis[i])
```

```python
cov=zeros((5,5),float)
for i in range(5):
    for j in range(5):
        cov[i,j]=dot(basis[i],basis[j])
        
figure()
imshow(cov)
```
