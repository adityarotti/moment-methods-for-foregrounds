```python
%pylab notebook

h=6.62607004e-34 #Plancks constant #m2 kg / s
k=1.38064852e-23 #Boltzmann constant #m2 kg s-2 K-1
```

```python
def gram_schmidt(vectors):
    basis = []
    for v in vectors:
        w = v - np.sum( np.dot(v,b)*b  for b in basis )
        #norm=np.linalg.norm(w) ; print norm
        norm=np.sqrt(np.dot(w,w)) #; print norm
        if (norm > 1e-16):  
            basis.append(w/norm)
    return np.array(basis)
```

```python
def y1(x):
    return x*exp(x)/(exp(x)-1.)

def y2(x):
    return y1(x)*x*cosh(0.5*x)/sinh(0.5*x)

def y3(x):
    return y1(x)*x*x*(cosh(x)+2)/(cosh(x)-1)

def y4(x):
    return y2(x)*0.5*x*x*(cosh(x)+5)/(sinh(0.5*x)**2.)

def y5(x):
    return y1(x)*(x**4.)*(33.+26*cosh(x)+cosh(2.*x))/(8.*sinh(0.5*x)**4.)

def f_mbb(nu,T,alpha,A,p22,p23,p33,p222,p223,p233,p333):
    x=h*nu*1e9/k/T
    nu0=k*T/h/1e9
    I0=A*(((nu/nu0)**alpha)*(nu**3.))/(exp(x)-1.)
    temp=I0*(1. + 0.5*p22*(log(nu/nu0))**2. + p23*log(nu/nu0)*y1(x)+ 0.5*p33*y2(x))
    temp = temp +  I0*(p222*(log(nu/nu0)**3.)/6. + 0.5*p223*(log(nu/nu0)**2.)*y1(x) + 0.5*p233*log(nu/nu0)*y2(x) + p333*y3(x)/6.)
    return temp
```

```python
def fn_v0(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=((nu/nu0)**alpha)*(nu**3.)/(exp(x)-1.)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v1(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*log(nu/nu0)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v2(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*y1(x)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v3(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*(log(nu/nu0)**2.)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v4(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*log(nu/nu0)*y1(x)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v5(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*y2(x)/(nu**sl)
    Inu=Inu/max(Inu)
    return Inu

def fn_v6(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*(log(nu/nu0)**3.)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v7(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*y1(x)*(log(nu/nu0)**2)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v8(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*y2(x)*log(nu/nu0)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v9(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*y3(x)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v10(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*(log(nu/nu0)**4)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v11(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*(log(nu/nu0)**3)*y1(x)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v12(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*(log(nu/nu0)**2)*y2(x)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v13(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*(log(nu/nu0)**1.)*y3(x)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v14(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*y4(x)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v15(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*(log(nu/nu0)**4.)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

def fn_v16(nu,T,alpha,sl=0):
    x=h*nu*1e9/(k*T)
    nu0=k*T/h/1e9
    Inu=fn_v0(nu,T,alpha,sl=sl)*(log(nu/nu0)**5.)/(nu**sl)
    Inu=Inu/max(abs(Inu))
    return Inu

fn_dir=[fn_v0,fn_v1,fn_v2,fn_v3,fn_v4,fn_v5,fn_v6,fn_v7,fn_v8,fn_v9,fn_v10,fn_v11,fn_v12,fn_v13,fn_v14]#,fn_v15,fn_v16]
```

```python
def mbb(nu,T,A,alpha,nu0=1.):
    x=h*nu*1e9/(k*T)
    Inu=A*((nu/nu0)**alpha)*(nu**3.)/(exp(x)-1.)
    return Inu

def multitemp_mbb_spectra(nu,T1,A1,alpha1,T2,A2,alpha2,nu0=1.):
    Inu=mbb(nu,T1,A1,alpha1,nu0=nu0) + mbb(nu,T2,A2,alpha2,nu0=nu0)
    return Inu
```
