```python
import healpy as h
import numpy as np
%pylab notebook
```

```python
datapath="../data/Planck/"
fname_prefix="HFI_SkyMap_"
fname_suffix="_2048_R2.02_nominal.fits"

nsideout=256
fname_suffix_out="_"+ str(nsideout) +"_R2.02_nominal.fits"
npix=h.nside2npix(nsideout)
freq=[100,143,217,353,545,857]

for i in range(np.size(freq)):
    filename=datapath + fname_prefix + str(freq[i]) + fname_suffix
    filename_out=datapath + fname_prefix + str(freq[i]) + fname_suffix_out
    data=h.read_map(filename,verbose=False,field=[0,2])
    hfi=h.ud_grade(data[0],nsideout)
    hfi_cov=h.ud_grade(data[1],nsideout,power=2)
    h.write_map(filename_out,[hfi,hfi_cov],overwrite=True)
```
