```python
%pylab notebook
import time
import sys  ; sys.path.append("../")
import sympy as sp
import numpy as np
from scipy.integrate import fixed_quad

from modules import  analytic_sed
from modules import constants as cnst

#from sympy.utilities.autowrap import ufuncify
```

```python
an=analytic_sed.analytic_sed()
an.create_sym_fn_dir(5)
nu=linspace(30,3000,500)

T=15 ; slope=1.
nu0=cnst.boltzman_const*T/cnst.planck_const/cnst.ghz2hz 
c0=cnst.planck_const*cnst.ghz2hz/cnst.boltzman_const

def dot_prod(v1,v2,xmin,xmax):
    return fixed_quad(lambda x: v1(x)*v2(x),xmin,xmax,n=80)[0]
    #return quad(lambda x: v1(x)*v2(x),xmin,xmax)[0]
```

```python
sym_vec={}
#x=sp.symbols("x")
for i in range(len(an.sym_fn_dir)):
    fn=an.sym_fn_dir[i]
    sym_vec[i]=fn.subs([(an.beta,1./T),(an.s,slope),(an.nu0,nu0),(an.c0,c0)])
```

```python
vec={}
for i in range(len(an.sym_fn_dir)):
    vec[i]=sp.lambdify(an.nu,sym_vec[i],modules="numpy")
    #vec[i]=ufuncify(an.nu,sym_vec[i])
    
figure()
for j in range(len(vec)):
    plot(nu,vec[j](nu))
```

```python
def get_gs_basis(sym_vec,xmin,xmax):
    sym_basis={}
    cnt=0
    for i in range(len(sym_vec)):    
        #print "i=",i
        vi=sp.lambdify(an.nu,sym_vec[i],modules="numpy")
        #vi=ufuncify(an.nu,sym_vec[i],backend="numpy")
        bdotv=[]
        for j in range(len(sym_basis)):
            bj=sp.lambdify(an.nu,sym_basis[j],modules="numpy")
            #bj=ufuncify(an.nu,sym_basis[j],backend="numpy")
            numerator=dot_prod(vi,bj,xmin,xmax)
            denominator=dot_prod(bj,bj,xmin,xmax) #; print denominator
            bdotv=append(bdotv,numerator/denominator)
        fn=sp.lambdify(an.nu,sym_vec[i]-sum(bdotv[j]*sym_basis[j] for j in range(len(bdotv))),modules="numpy")
        norm=dot_prod(fn,fn,xmin,xmax) #; print sqrt(norm)
        sym_basis[i]=(sym_vec[i]-sum(bdotv[j]*sym_basis[j] for j in range(len(bdotv))))/sqrt(norm)
        #print i #,len(sym_vec)#,sym_basis[i]
    return sym_basis

start = time.time()
sym_basis_o=get_gs_basis(sym_vec,min(nu),max(nu))
for i in range(3):
    sym_basis=get_gs_basis(sym_basis_o,min(nu),max(nu))
end = time.time()
print end-start
```

```python
figure()
for i in range(len(sym_basis_o)):
    fn=sp.lambdify(an.nu,sym_basis_o[i],modules="numpy")
    plot(nu,fn(nu),label=str(i))
#semilogy()
#legend(loc=0)
ylim(-0.051,0.051)
```

```python
cov=zeros((len(sym_basis),len(sym_basis)),float)

for i in xrange(len(sym_basis)):
    print i
    for j in xrange(i+1):
        bi=sp.lambdify(an.nu,sym_basis[i],modules="numpy")
        bj=sp.lambdify(an.nu,sym_basis[j],modules="numpy")
        cov[i,j]=dot_prod(bi,bj,min(nu),max(nu))
        cov[j,i]=cov[i,j]
```
